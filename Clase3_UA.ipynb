```python
# prompt: tabla de frecuecia para NRC y diagrama de caja

import pandas as pd
import matplotlib.pyplot as plt
# Calculate the frequency table for 'NRC'
nrc_counts = df['NRC'].value_counts().reset_index()
nrc_counts.columns = ['NRC', 'Frecuencia']

# Add columns for relative frequency and cumulative frequency
nrc_counts['Frecuencia Relativa'] = nrc_counts['Frecuencia'] / nrc_counts['Frecuencia'].sum()
nrc_counts['Frecuencia Acumulada'] = nrc_counts['Frecuencia'].cumsum()
nrc_counts['Frecuencia Relativa Acumulada'] = nrc_counts['Frecuencia Acumulada'] / nrc_counts['Frecuencia'].sum()

print("Tabla de Frecuencia para NRC:")
print(nrc_counts)

# Create a box plot for 'NRC' (assuming 'NRC' represents a numerical value or you want to see the distribution of another numerical column grouped by NRC)
# If NRC is categorical and you want to see the distribution of a numerical variable (e.g., 'PARCIAL 1') for each NRC:
plt.figure(figsize=(12, 6))
sns.boxplot(x='NRC', y='PARCIAL 1', data=df)
plt.title('Diagrama de Caja de PARCIAL 1 por NRC')
plt.xlabel('NRC')
plt.ylabel('PARCIAL 1')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()

# If 'NRC' is treated as a numerical variable and you want its box plot:
# Convert NRC to numeric first if necessary
# df['NRC_numeric'] = pd.to_numeric(df['NRC'], errors='coerce')
# plt.figure(figsize=(8, 6))
# sns.boxplot(y=df['NRC_numeric'])
# plt.title('Diagrama de Caja para NRC')
# plt.ylabel('NRC')
# plt.grid(True)
# plt.show()
```
